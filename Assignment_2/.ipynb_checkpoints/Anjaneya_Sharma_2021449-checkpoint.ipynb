***Part 2 A***

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# Load the data
data = scipy.io.loadmat('Assignment2_2A_NDM_2024.mat', squeeze_me=True)['NDM_Assignment2']

# Constants
start_threshold = 300
decision_threshold = 600
num_bins = 20
bin_size = 50  # Each bin is 50 ms

# Function to calculate reaction time (RT) for a trial
def calculate_rt(trial_data):
    for time_index, value in enumerate(trial_data):
        if value >= decision_threshold:
            return time_index  # Returns the time index where threshold is first reached
    return np.nan  # If threshold is never reached

# Initialize lists to store mean RT for all participants in both conditions
mean_rt_condition_1 = []
mean_rt_condition_2 = []

# Process data for each condition
for condition in range(2):  # Condition 1 and Condition 2
    mean_rt_per_participant = []
    for participant_data in data[:, condition]:  # Each participant's data for a condition
        # Calculate RT for all trials
        rt_trials = [calculate_rt(trial) for trial in participant_data]
        rt_trials = np.array(rt_trials) - start_threshold  # Adjust by the start threshold

        # Exclude trials where threshold was never reached (resulted in NaN)
        rt_trials = rt_trials[~np.isnan(rt_trials)]
        
        # Calculate mean RT for this participant
        mean_rt_per_participant.append(np.mean(rt_trials))  # Store mean RT for this participant

        
    
    # Append to respective condition list
    if condition == 0:
        mean_rt_condition_1.extend(mean_rt_per_participant)
    else:
        mean_rt_condition_2.extend(mean_rt_per_participant)

# Plot histograms for both conditions
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.histplot(mean_rt_condition_1, bins=num_bins, color='skyblue', edgecolor='black', kde=False, ax=axes[0])
mean_rt_1 = np.mean(mean_rt_condition_1)
axes[0].axvline(mean_rt_1, color='red', linestyle='dashed', linewidth=1)
axes[0].set_title(f'High Coherence\nMean RT = {mean_rt_1:.2f} ms')
axes[0].set_xlabel('Reaction Time (ms)')
axes[0].set_ylabel('Frequency (Number of Participants)')

sns.histplot(mean_rt_condition_2, bins=num_bins, color='skyblue', edgecolor='black', kde=False, ax=axes[1])
mean_rt_2 = np.mean(mean_rt_condition_2)
axes[1].axvline(mean_rt_2, color='red', linestyle='dashed', linewidth=1)
axes[1].set_title(f'Low Coherence\nMean RT = {mean_rt_2:.2f} ms')
axes[1].set_xlabel('Reaction Time (ms)')
axes[1].set_ylabel('Frequency (Number of Participants)')

plt.suptitle("Reaction Time Distributions in Random Dot Motion Task")
plt.tight_layout()
plt.show()

# Perform Mann-Whitney U test
u_stat, p_value = stats.mannwhitneyu(mean_rt_condition_1, mean_rt_condition_2, alternative='two-sided')

# Print results of the Mann-Whitney U test
print("Mann-Whitney U test Results:")
print(f"U Statistic: {u_stat:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
# Load the data
data = loadmat('Assignment2_2A_NDM_2024.mat')['NDM_Assignment2']
n_participants, n_conditions = data.shape ######## check

# Initialize dictionaries to store reaction times (RTs) for each condition
rts = {0: [], 1: []}     ########## check

# Calculate RT for each trial in each condition
for condition in range(2):
    for participant in range(n_participants):
        participant_data = data[participant, condition]
        trial_rts = []       ##### checkkkkk
        
        # Calculate RT for each trial
        for trial in participant_data:
            threshold_indices = np.where(trial >= 600)[0]
            # Check if the threshold is reached in the trial
            if len(threshold_indices) > 0:
                rt = threshold_indices[0]  # First time index where threshold is reached # checkkk
            else:
                rt = np.nan  # If threshold is never reached
            trial_rts.append(rt)
        
        rts[condition].append(trial_rts)

# Calculate mean RT for each participant in each condition
mean_rt_cond1 = np.nanmean(rts[0], axis=1)
mean_rt_cond2 = np.nanmean(rts[1], axis=1)

# Perform Shapiro-Wilk test for normality
_, p_val_1 = stats.shapiro(mean_rt_cond1)
_, p_val_2 = stats.shapiro(mean_rt_cond2)

# Select appropriate statistical test based on normality
if p_val_1 > 0.05 and p_val_2 > 0.05:
    statistic, p_value = stats.ttest_ind(mean_rt_cond1, mean_rt_cond2, nan_policy='omit')
    test_name = "Independent t-test"
else:
    statistic, p_value = stats.mannwhitneyu(mean_rt_cond1, mean_rt_cond2)
    test_name = "Mann-Whitney U test"

# Print statistical results      #### refractor the results printing part
print("\nStatistical Analysis Results:")
print(f"Test used: {test_name}")
print(f"Test statistic: {statistic}")
print(f"p-value: {p_value}")
print(f"\nCondition 1 (High Coherence): Mean RT = {np.nanmean(mean_rt_cond1):.1f} ms")
print(f"\nCondition 2 (Low Coherence): Mean RT = {np.nanmean(mean_rt_cond2):.1f} ms")

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8), facecolor='white')
bins = np.linspace(0, 1000, 21)
condition_names = ['High Coherence', 'Low Coherence']

# Basic plot for each condition
for condition, (ax, name) in enumerate(zip([ax1, ax2], condition_names)):
    participant_means = np.nanmean(rts[condition], axis=1)
    mean_rt = np.nanmean(participant_means)
    
    ax.hist(participant_means, bins=bins, color='green', edgecolor='black')
    ax.axvline(mean_rt, color='red', linestyle='--', linewidth=1.5, label=f'Mean RT: {mean_rt:.1f} ms')
    
    ax.set_xlabel('Reaction Time (ms)', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_title(f'{name} (Mean RT = {mean_rt:.1f} ms)', fontsize=14)
    ax.legend(fontsize=10)

# Main title and caption
plt.suptitle('Distribution of reaction times in Random Dot Motion Task', fontsize=16, color='black')


# Show plot without extra layout adjustments
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz
import scipy.stats as stats


data_xl = pd.read_excel('Assignment2-2B-NDM-2024.xlsx', index_col = False)

# Load the Excel file
file_path = 'Assignment2-2B-NDM-2024.xlsx'
xls = pd.ExcelFile(file_path)

# Dictionary to store AUC for each condition
auc_dict = {}

# Iterate over each sheet to calculate AUC
for cnt, sheet_name in enumerate(xls.sheet_names, start=1):
    # Read the sheet into a DataFrame, skipping the first two rows
    df = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=1)
    
    # Initialize a list to store AUCs for each participant
    condition_aucs = []

    # Different column indexing for the first sheet (has an extra column)
    if cnt == 1:
        # Separate HT and FA columns for the first sheet (starting from column index 1)
        ht_cols = df.iloc[:, 1::3]
        fa_cols = df.iloc[:, 2::3]
    else:
        # Separate HT and FA columns for other sheets
        ht_cols = df.iloc[:, ::3]
        fa_cols = df.iloc[:, 1::3]
    
    # Ensure columns are numeric
    ht_cols = ht_cols.apply(pd.to_numeric, errors='coerce')
    fa_cols = fa_cols.apply(pd.to_numeric, errors='coerce')
    
    # Rename columns of fa_cols to match ht_cols for easy referencing
    fa_cols.columns = ht_cols.columns
    
    # Calculate AUC for each participant
    for participant in range(ht_cols.shape[1]):
        ht = ht_cols.iloc[:, participant].values
        fa = fa_cols.iloc[:, participant].values

        # Sort fa and reorder ht accordingly for accurate AUC calculation
        sorted_indices = np.argsort(fa)
        fa_sorted = fa[sorted_indices]
        ht_sorted = ht[sorted_indices]
        
        # Calculate AUC using trapezoidal integration
        auc_value = trapz(ht_sorted, fa_sorted)
        condition_aucs.append(auc_value)
    
    # Store the AUCs for this condition
    auc_dict[sheet_name] = condition_aucs  

# Create a DataFrame from the dictionary
auc_df = pd.DataFrame(auc_dict, columns=['Cond1', 'Cond2', 'Cond3', 'Cond4'])

# Display the DataFrame and mean AUC for each condition
print(auc_df)
for i in range(4):
    print(f"Mean AUC for Cond{i+1}: {auc_df.iloc[:, i].mean()}")

# Plotting box and whisker plots for each condition
plt.figure(figsize=(12, 6))
auc_df.boxplot()
plt.title('Box and Whisker Plot of AUC Values for Each Experimental Condition')
plt.xlabel('Experimental Conditions')
plt.ylabel('AUC (Accuracy)')
plt.grid(False)
plt.show()

# Perform Friedman's test
friedman_stat, p_value = stats.friedmanchisquare(
    auc_df['Cond1'], auc_df['Cond2'], auc_df['Cond3'], auc_df['Cond4']
)

# Print results of Friedman's test
print(f"Friedman's Chi-square statistic: {friedman_stat}")
print(f"p-value: {p_value}")
```
